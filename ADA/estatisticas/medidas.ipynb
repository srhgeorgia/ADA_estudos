# üìä Medidas de Posi√ß√£o e Dispers√£o
Este notebook apresenta exemplos pr√°ticos e explica√ß√µes sobre as principais medidas **estat√≠sticas descritivas**.

As medidas s√£o divididas em dois grupos:
- **Medidas de posi√ß√£o (tend√™ncia central)**: indicam onde os dados se concentram.
- **Medidas de dispers√£o (variabilidade)**: mostram o quanto os dados variam.

## üß≠ Medidas de Posi√ß√£o

As principais medidas de posi√ß√£o s√£o:
- **Moda**: valor que mais aparece.
- **M√©dia**: soma dos valores dividida pela quantidade de elementos.
- **Mediana**: valor central da s√©rie ordenada.
- **Quartis**: dividem os dados em quatro partes.
- **Percentis**: dividem os dados em cem partes.

```python

import numpy as np
import pandas as pd

# Conjunto de dados
dados = [10, 12, 13, 13, 14, 16, 18, 21, 21, 21, 25, 30]
serie = pd.Series(dados)

# C√°lculos
moda = serie.mode().tolist()
media = serie.mean()
mediana = serie.median()
quartis = serie.quantile([0.25, 0.5, 0.75])
percentil_90 = serie.quantile(0.9)

print("Moda:", moda)
print("M√©dia:", media)
print("Mediana:", mediana)
print("\nQuartis:")
print(quartis)
print("\nPercentil 90:", percentil_90)
```

### üîç Interpreta√ß√£o
- **Moda**: o valor mais frequente.
- **M√©dia**: o ponto de equil√≠brio dos dados.
- **Mediana**: o valor central (metade dos dados est√° acima, metade abaixo).
- **Quartis**: mostram a distribui√ß√£o em quatro partes.
- **Percentil 90**: indica o valor abaixo do qual est√£o 90% dos dados.

## üìâ Medidas de Dispers√£o

As medidas de dispers√£o mostram **o quanto os dados est√£o espalhados**.
- **Amplitude total**: diferen√ßa entre o maior e o menor valor.
- **Vari√¢ncia**: m√©dia dos quadrados dos desvios em rela√ß√£o √† m√©dia.
- **Desvio padr√£o**: raiz quadrada da vari√¢ncia.

```python

amplitude = serie.max() - serie.min()
variancia = serie.var()
desvio_padrao = serie.std()

print("Amplitude total:", amplitude)
print("Vari√¢ncia:", round(variancia, 2))
print("Desvio padr√£o:", round(desvio_padrao, 2))
```

### üìà Interpreta√ß√£o
- **Amplitude** mostra a diferen√ßa total dos valores.
- **Vari√¢ncia** e **desvio padr√£o** indicam o quanto os dados se afastam da m√©dia.
- **Quanto maior o desvio padr√£o**, maior a dispers√£o dos dados.

## üìä Visualiza√ß√£o da Distribui√ß√£o

```python

import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(dados, bins=6, color="skyblue", edgecolor="black")
plt.axvline(media, color="red", linestyle="--", label=f"M√©dia = {media:.1f}")
plt.axvline(mediana, color="green", linestyle="--", label=f"Mediana = {mediana:.1f}")
plt.title("Distribui√ß√£o dos Dados ‚Äî Medidas de Posi√ß√£o")
plt.xlabel("Valor")
plt.ylabel("Frequ√™ncia")
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

‚ú® **Conclus√£o:**
- As **medidas de posi√ß√£o** resumem onde os dados se concentram.
- As **medidas de dispers√£o** mostram o quanto eles se afastam.
- Juntas, ajudam a entender a estrutura e variabilidade de qualquer conjunto de dados.
