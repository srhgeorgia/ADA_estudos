# ðŸ“Š DistribuiÃ§Ã£o de FrequÃªncias â€” Exemplo PrÃ¡tico
Este notebook demonstra a construÃ§Ã£o de uma **tabela de distribuiÃ§Ã£o de frequÃªncias**, usando como exemplo os **pesos dos alunos de uma escola**.

Inclui os seguintes cÃ¡lculos:
- Intervalo de classes
- FrequÃªncia absoluta
- Ponto mÃ©dio
- FrequÃªncia relativa
- FrequÃªncia acumulada
- FrequÃªncia relativa acumulada

```python

import pandas as pd
import matplotlib.pyplot as plt

# Dados de exemplo â€” baseados na imagem
intervalos = ["41â€“45", "45â€“49", "49â€“53", "53â€“57", "57â€“61"]
freq_absoluta = [7, 3, 4, 1, 5]

# CriaÃ§Ã£o do DataFrame
df = pd.DataFrame({
    "Intervalo de classes": intervalos,
    "FrequÃªncia absoluta": freq_absoluta
})

# Total de observaÃ§Ãµes
total = sum(freq_absoluta)

# Ponto mÃ©dio = (limite inferior + limite superior) / 2
df["Ponto mÃ©dio"] = [43, 47, 51, 55, 59]

# FrequÃªncia relativa
df["FrequÃªncia relativa"] = df["FrequÃªncia absoluta"] / total

# FrequÃªncia acumulada
df["FrequÃªncia acumulada"] = df["FrequÃªncia absoluta"].cumsum()

# FrequÃªncia relativa acumulada
df["FrequÃªncia relativa acumulada"] = df["FrequÃªncia relativa"].cumsum()

display(df)
```

## ðŸ“‹ InterpretaÃ§Ã£o da tabela

- A **frequÃªncia absoluta** mostra quantos alunos estÃ£o em cada intervalo de peso.
- O **ponto mÃ©dio** Ã© a mÃ©dia entre os limites inferior e superior do intervalo.
- A **frequÃªncia relativa** mostra a proporÃ§Ã£o de alunos em cada faixa.
- A **frequÃªncia acumulada** soma progressivamente as frequÃªncias absolutas.
- A **frequÃªncia relativa acumulada** soma progressivamente as proporÃ§Ãµes, terminando sempre em 1.0 (ou 100%).

## ðŸ“Š GrÃ¡fico de barras â€” FrequÃªncia absoluta

```python

plt.figure(figsize=(8,5))
plt.bar(df["Intervalo de classes"], df["FrequÃªncia absoluta"], color="royalblue")
plt.title("DistribuiÃ§Ã£o de FrequÃªncias â€” Peso dos alunos")
plt.xlabel("Intervalo de classes (kg)")
plt.ylabel("FrequÃªncia absoluta")
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.show()
```

## ðŸ“ˆ GrÃ¡fico de linha â€” FrequÃªncia relativa acumulada

```python

plt.figure(figsize=(8,5))
plt.plot(df["Ponto mÃ©dio"], df["FrequÃªncia relativa acumulada"], marker="o", color="darkorange")
plt.title("FrequÃªncia Relativa Acumulada â€” Curva de Ogiva")
plt.xlabel("Ponto mÃ©dio (kg)")
plt.ylabel("FrequÃªncia relativa acumulada")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

âœ¨ **ConclusÃµes:**
- 35% dos alunos pesam entre 41â€“45 kg.
- 75% dos alunos pesam atÃ© 57 kg.
- A frequÃªncia relativa acumulada sempre atinge 1, indicando 100% da amostra.
