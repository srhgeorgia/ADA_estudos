# 游늵 Exemplos de Gr치ficos em Python (Matplotlib)
Este notebook mostra exemplos pr치ticos dos principais tipos de gr치ficos utilizados na an치lise de dados.

### Tipos de gr치ficos inclu칤dos:
- Gr치fico de linhas
- Gr치fico de colunas
- Gr치fico de barras
- Gr치fico de setores (pizza)
- Histograma
- Diagrama de caixas (boxplot)
- Gr치fico de dispers칚o (scatter)


```python

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Dados b치sicos
anos = np.arange(2015, 2025)
vendas = [12, 15, 14, 18, 20, 25, 22, 28, 26, 30]

df = pd.DataFrame({"Ano": anos, "Vendas": vendas})
display(df)
```

## 游늳 1. Gr치fico de Linhas

```python

plt.figure(figsize=(8,5))
plt.plot(df["Ano"], df["Vendas"], marker="o", color="darkorange")
plt.title("Gr치fico de Linhas - Vendas ao longo dos anos")
plt.xlabel("Ano")
plt.ylabel("Vendas (em milh칫es)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

## 游릱 2. Gr치fico de Colunas (barras verticais)

```python

plt.figure(figsize=(8,5))
plt.bar(df["Ano"], df["Vendas"], color="royalblue")
plt.title("Gr치fico de Colunas - Vendas por Ano")
plt.xlabel("Ano")
plt.ylabel("Vendas (em milh칫es)")
plt.show()
```

## 游릳 3. Gr치fico de Barras (horizontais)

```python

plt.figure(figsize=(8,5))
plt.barh(df["Ano"], df["Vendas"], color="seagreen")
plt.title("Gr치fico de Barras - Vendas por Ano")
plt.xlabel("Vendas (em milh칫es)")
plt.ylabel("Ano")
plt.show()
```

## 游볺 4. Gr치fico de Setores (Pizza)

```python

fatias = [30, 25, 20, 15, 10]
categorias = ["A", "B", "C", "D", "E"]
plt.figure(figsize=(6,6))
plt.pie(fatias, labels=categorias, autopct='%1.1f%%', startangle=90, colors=plt.cm.Pastel1.colors)
plt.title("Gr치fico de Setores - Distribui칞칚o por Categoria")
plt.show()
```

## 游늵 5. Histograma

```python

dados = np.random.normal(50, 10, 200)
plt.figure(figsize=(8,5))
plt.hist(dados, bins=10, color="orchid", edgecolor="black")
plt.title("Histograma - Distribui칞칚o de Valores")
plt.xlabel("Faixas de valor")
plt.ylabel("Frequ칡ncia")
plt.show()
```

## 游닍 6. Diagrama de Caixas (Boxplot)

```python

dados1 = np.random.normal(50, 10, 100)
dados2 = np.random.normal(60, 15, 100)

plt.figure(figsize=(8,5))
plt.boxplot([dados1, dados2], labels=["Amostra 1", "Amostra 2"], patch_artist=True,
            boxprops=dict(facecolor="lightblue"))
plt.title("Diagrama de Caixas - Compara칞칚o de Amostras")
plt.ylabel("Valores")
plt.show()
```

## 丘 7. Gr치fico de Dispers칚o (Scatter Plot)

```python

x = np.random.rand(50) * 100
y = x + np.random.randn(50) * 10

plt.figure(figsize=(8,5))
plt.scatter(x, y, color="tomato")
plt.title("Gr치fico de Dispers칚o - Rela칞칚o entre X e Y")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

- Cada tipo de gr치fico ajuda a visualizar padr칫es diferentes.
- Gr치ficos de linha e barras mostram evolu칞칚o e compara칞칚o.
- Setores e histogramas mostram propor칞칫es e distribui칞칫es.
- Boxplots e dispers칫es ajudam na an치lise estat칤stica e correla칞칚o.
