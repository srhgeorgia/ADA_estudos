# üìò Tipos de Dados ‚Äî Exemplos em Python (Jupyter)
Este notebook demonstra, de forma pr√°tica, como funcionam os **tipos de dados**: categorizados (qualitativos) e num√©ricos (quantitativos).

## Estrutura geral:
- Categorizados (Qualitativos)
  - Nominais
  - Ordinais
- Num√©ricos (Quantitativos)
  - Discretos
  - Cont√≠nuos


## üß© 1. Dados Categorizados (Qualitativos)

### a) Nominais ‚Äî sem ordem natural

```python

# Exemplo de dados nominais
cores = ["Vermelho", "Azul", "Verde", "Azul", "Vermelho"]

from collections import Counter
contagem_cores = Counter(cores)

print("Frequ√™ncia das cores:")
print(contagem_cores)
```

### b) Ordinais ‚Äî com ordem natural

```python

avaliacoes = ["Ruim", "Regular", "Bom", "√ìtimo", "Bom", "Regular"]

ordem = {"Ruim": 1, "Regular": 2, "Bom": 3, "√ìtimo": 4}
avaliacoes_numericas = [ordem[a] for a in avaliacoes]

import numpy as np
media = np.mean(avaliacoes_numericas)

print("M√©dia das avalia√ß√µes (escala 1 a 4):", round(media, 2))
```

## üî¢ 2. Dados Num√©ricos (Quantitativos)

### a) Discretos ‚Äî valores inteiros, cont√°veis

```python

alunos_por_turma = [20, 22, 18, 25, 20, 19]

print("M√©dia:", np.mean(alunos_por_turma))
print("M√°ximo:", np.max(alunos_por_turma))
print("M√≠nimo:", np.min(alunos_por_turma))
```

### b) Cont√≠nuos ‚Äî valores fracion√°rios, medidos

```python

alturas = [1.65, 1.72, 1.81, 1.59, 1.68, 1.75]

print("M√©dia:", np.mean(alturas))
print("Desvio padr√£o:", np.std(alturas))
```

## üìä 3. Visualiza√ß√£o dos Dados

```python

import matplotlib.pyplot as plt

plt.bar(contagem_cores.keys(), contagem_cores.values(), color=['red', 'blue', 'green'])
plt.title("Frequ√™ncia de Cores (Nominais)")
plt.xlabel("Cor")
plt.ylabel("Frequ√™ncia")
plt.show()
```

‚ú® **Conclus√£o:**
- Nominais e Ordinais representam **categorias**.
- Discretos e Cont√≠nuos representam **valores num√©ricos**.
- √â importante distinguir esses tipos para aplicar corretamente an√°lises estat√≠sticas e visualiza√ß√µes.
