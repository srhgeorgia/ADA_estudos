Este notebook demonstra, com exemplos pr√°ticos, como construir e compreender tabelas.

## Estrutura de uma tabela:
- **T√≠tulo:** explica o conte√∫do da tabela
- **Corpo:** conjunto de linhas e colunas com as informa√ß√µes
- **Cabe√ßalho:** parte superior da tabela (nomes das colunas)
- **Coluna indicadora:** identifica o conte√∫do das linhas (√≠ndice)
- **Casa ou c√©lula:** espa√ßo onde h√° um valor (interse√ß√£o de linha e coluna)
- **Linha:** c√©lulas organizadas horizontalmente
- **Coluna:** c√©lulas organizadas verticalmente
- **Rodap√©:** elemento complementar (fonte, observa√ß√£o, nota etc.)

## üìã Exemplo 1 ‚Äî Criando uma tabela simples com pandas

```python

import pandas as pd

# Dados
dados = {
    "Nome": ["Ana", "Bruno", "Carla", "Diego"],
    "Idade": [23, 29, 31, 26],
    "Cidade": ["Curitiba", "S√£o Paulo", "Florian√≥polis", "Belo Horizonte"]
}

# Cria√ß√£o da tabela
tabela = pd.DataFrame(dados)

# Exibi√ß√£o
print("T√≠tulo: Tabela de Pessoas\n")
display(tabela)
```

Neste exemplo:
- **T√≠tulo:** Tabela de Pessoas
- **Cabe√ßalho:** Nome, Idade, Cidade
- **Corpo:** dados de cada pessoa
- **C√©lulas:** valores individuais (ex: ‚ÄúAna‚Äù, ‚Äú23‚Äù)
- **Coluna indicadora:** √≠ndice (0, 1, 2, 3)

## üìé Exemplo 2 ‚Äî Adicionando um rodap√© (fonte de dados)

```python

print("T√≠tulo: Tabela de Pessoas\n")
display(tabela)
print("\nRodap√©: Fonte - Dados fict√≠cios criados para fins did√°ticos.")
```
